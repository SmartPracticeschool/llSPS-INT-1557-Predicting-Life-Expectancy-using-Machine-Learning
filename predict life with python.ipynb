```python
data = pd.concat([data, status], axis = 1)
data = data.drop(['Status'], axis=1)
data.rename(columns = {'Deloping' : '0', 'Developed' : 1})
```

```python
x = data.drop(['Life expectancy '], axis=1)
y = data['Life expectancy ']
```

```python
from sklearn.model_selection import train_test_split
train_x,valid_x,train_y,valid_y = train_test_split(x,y, random_state = 101, shuffle=False)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(random_state=95)
```

```python
model.fit(train_x,train_y)
```

```python
pred=model.predict(valid_x)
```

```python
model.score(valid_x, valid_y)
```

```python
from sklearn import metrics
```

```python
print('MAE:', metrics.mean_absolute_error(valid_y, pred))
print('MSE:', metrics.mean_squared_error(valid_y, pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(valid_y, pred)))
```
